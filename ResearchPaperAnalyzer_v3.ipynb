```python
import torch
from transformers import BitsAndBytesConfig
from llama_index.core.prompts import PromptTemplate
from llama_index.llms.huggingface import HuggingFaceLLM

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)


def messages_to_prompt(messages):
  prompt = ""
  for message in messages:
    if message.role == 'system':
      prompt += f"<|system|>\n{message.content}</s>\n"
    elif message.role == 'user':
      prompt += f"<|user|>\n{message.content}</s>\n"
    elif message.role == 'assistant':
      prompt += f"<|assistant|>\n{message.content}</s>\n"

  # ensure we start with a system prompt, insert blank if needed
  if not prompt.startswith("<|system|>\n"):
    prompt = "<|system|>\n</s>\n" + prompt

  # add final assistant prompt
  prompt = prompt + "<|assistant|>\n"

  return prompt


llm = HuggingFaceLLM(
    model_name="HuggingFaceH4/zephyr-7b-beta",
    tokenizer_name="HuggingFaceH4/zephyr-7b-beta",
    query_wrapper_prompt=PromptTemplate("<|system|>\n</s>\n<|user|>\n{query_str}</s>\n<|assistant|>\n"),
    context_window=3900,
    max_new_tokens=512,
    model_kwargs={"quantization_config": quantization_config},
    # tokenizer_kwargs={},
    generate_kwargs={"temperature": 0.7, "top_k": 50, "top_p": 0.95},
    messages_to_prompt=messages_to_prompt,
    device_map="auto",
)
```

```python
import requests

def download_pdf(url, file_path):
    response = requests.get(url, stream=True)

    if response.status_code == 200:
        with open(file_path, "wb") as file:
            for chunk in response.iter_content(chunk_size=1024):
                file.write(chunk)
        print("PDF downloaded successfully to", file_path)
    else:
        raise Exception(f"Failed to download PDF: {response.status_code}")
```

```python
import urllib.parse
from typing import Optional
import json
import requests
from llama_index.core.schema import Document
from llama_index.core.tools.tool_spec.base import BaseToolSpec
from main_content_extractor import MainContentExtractor

QUERY_URL_TMPL = ("https://www.googleapis.com/customsearch/v1?key={key}&cx={engine}&q={query}&num=2")


class GoogleSearchToolSpec(BaseToolSpec):
    """Google Search tool spec."""

    spec_functions = ["google_search"]

    def __init__(self, key: str, engine: str) -> None:
        """Initialize with parameters."""
        self.key = key
        self.engine = engine

    def google_search(self, input: str):
        """
        Make a query to the Google search engine to receive a latest data on the query.
        Can also be used to get data of queries you dont have data about.


        Args:
            input (str): The query to be passed to Google search.

        Returns :
            Content of top 4 searches on the query in form of MarkDown

            Error 400 If failed, try again in that case.
        """
        url = QUERY_URL_TMPL.format(
            key=self.key, engine=self.engine, query=urllib.parse.quote_plus(input)
        )
        response = requests.get(url)
        print(response.status_code)
        print(response.content)
        res_json = json.loads(response.content.decode('utf-8'))
        # links = []
        content_final = ""

        headers = {
            "Content-Type":"text",
            "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36",
        }
        for items in res_json['items']:
          try:
            print('googling...')
            try:
              response = requests.get(items['link'], headers=headers)
              response.encoding = 'utf-8'
              content = response.text
              extracted_html = MainContentExtractor.extract(content)
              extracted_markdown = MainContentExtractor.extract(content, output_format="markdown")
              content_final += extracted_markdown
              break
            except:
              continue
          except Exception as e:
            print(e)
            continue

        return content_final
```

```python
import subprocess
from flask import Flask , jsonify ,request

from flask_cors import CORS
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader ,SummaryIndex
import json
from llama_index.core.tools import QueryEngineTool, ToolMetadata
from llama_index.core.agent import ReActAgent

app = Flask(__name__)
CORS(app)

url = ''
@app.route('/init_model',methods=['POST'])
def init_model():
    try:
      # global chat_engine
      global agent
      global url
      if 'agent' in globals():
        return jsonify({'success':False, 'message': 'Model Is ALREADY initialized'})
      else:
        url = request.form.get('url')
        file_path = "/content/data/research_paper.pdf"
        download_pdf(url, file_path)
        documents = SimpleDirectoryReader(input_dir="/content/data/").load_data()

        summary_index = SummaryIndex.from_documents(documents)
        vector_index = VectorStoreIndex.from_documents(documents)

        vector_tool = QueryEngineTool(
        vector_index.as_query_engine(verbose=True),
            metadata=ToolMetadata(
                name="vector_search",
                description="Useful for searching and explaining topics from the loaded research paper."
            )
        )
        summary_tool = QueryEngineTool(
            summary_index.as_query_engine(response_mode="tree_summarize",verbose=True),
            metadata=ToolMetadata(
                name="summary",
                description="Useful for summarizing text from the loaded research paper."
            )
        )
        google_tool_spec = GoogleSearchToolSpec(key="AIzaSyBQY1T-PSL2E1qkPRmoCC9CnMy5twbXk3A",engine="e273c4ae385d94161")

        google_tool = google_tool_spec.to_tool_list()[0]
        tools = [google_tool,vector_tool,summary_tool]
        agent = ReActAgent.from_tools(tools,verbose=True,memory=chat_memory)

        # chat_engine = vector_index.as_chat_engine(chat_mode="react", llm=llm, verbose=True,memory=chat_memory)

        return jsonify({'success': True, 'message': 'Model initialized successfully!'})

    except Exception as e:
      print(e)
      error_message = f"An error occurred: {str(e)}"

      if isinstance(e, requests.exceptions.RequestException):
          error_message = "Failed to download PDF. Check the URL or network connectivity."
      elif isinstance(e, FileNotFoundError):
          error_message = "Error accessing data directory. Check file paths."

      return jsonify({'success': False, 'message': error_message}), 500

@app.route('/chat', methods=['POST'])
def chat():
  """
  Returns:
      JSON response:
          - success (bool): True if successful, False otherwise.
          - response (str): The chat engine's response to the query.
          - message (str): Optional error message in case of issues.
  """
  try:
    query = request.form.get('query')
    response = agent.chat(query,)
    with open("/content/chat.json","w") as f:
      f.write(chat_store.json())
    return jsonify({'success': True, 'response': str(response)})

  except Exception as e:
    error_message = f"An error occurred: {str(e)}"

    return jsonify({'success': False, 'message': error_message}), 500

@app.route('/get_chat',methods=["GET"])
def get_chat():
  try:
    if not os.path.exists("/content/chat.json"):
      raise FileNotFoundError("chat.json file not found")

    with open("/content/chat.json", "r") as file:
      data = json.load(file)
    data = json.dumps(data)

    return jsonify({'success': True, 'data': data,'url':url})

  except Exception as e:
    # Handle general exceptions
    error_message = f"An error occurred: {str(e)}"

    # Return error response
    return jsonify({'success': False, 'message': error_message}), 500

def run_ngrok(domain, port):
    command = ["ngrok", "http", "--domain", domain, str(port)]
    process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

if __name__ == '__main__':
  domain = "herring-notable-physically.ngrok-free.app"
  port = 5000
  run_ngrok(domain,port)
  app.run(port=port)
```
